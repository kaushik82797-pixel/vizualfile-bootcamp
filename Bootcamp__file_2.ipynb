```python
students = [
    {"name": "Yash", "grade": 85},
    {"name": "Ram", "grade": 60},
    {"name": "Shyam", "grade": 99},
    {"name": "Krishna", "grade": 55},
    {"name": "Jay", "grade": 95}
]

for student in students:
    print("Student:", student["name"])
    print("Marks:", student["grade"])

    if student["grade"] >= 75:
        print("Performance: Strong")
    elif student["grade"] >= 50:
        print("Performance: Average")
    else:
        print("Performance: Needs Help")
```

```python
math_marks = np.array(df['math'])
print(math_marks)
```

```python
science_marks = np.array(df['science'])
print(science_marks)
print(science_marks.mean())
```

```python
df["Average"] = df[["math", "science", "english"]].mean(axis=1)

print(df)
```

```python
top_stu = df[df['Average'] > 85]
low_att = df[df['attendance_percent'] < 75]
mathura = df[df['city'] == "Mathura"]
```

```python
print(top_stu)
print(low_att)
print(mathura)
```
